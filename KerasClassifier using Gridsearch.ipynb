```python
import keras
```

```python
import keras.models as kmod
import keras.layers as klay
```

```python
import numpy as np
```

```python
arr = np.random.randint(0,2,(100,9))
```

```python
#dataset = np.loadtxt("pima-indians.csv", delimiter=",")
# split into input (X) and output (Y) variables
X = arr[:,0:8]
Y = arr[:,8]
# create model
model = kmod.Sequential()
model.add(klay.Dense(12, input_dim=8, activation='relu'))
model.add(klay.Dropout(0.1))
model.add(klay.Dense(8, activation='relu'))
model.add(klay.Dense(1, activation='sigmoid'))
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# Fit the model
model.fit(X, Y, epochs=100, batch_size=5)
# evaluate the model
scores = model.evaluate(X, Y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

# parameter tuning

```python
import keras.wrappers.scikit_learn as kwar
```

```python
def fn(optimizer):
    model = kmod.Sequential()
    model.add(klay.Dense(12, input_dim=8, activation='relu'))
    model.add(klay.Dropout(0.1))
    model.add(klay.Dense(8, activation='relu'))
    model.add(klay.Dense(1, activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])
    return model
```

```python
mod2 = kwar.KerasClassifier(fn)
```

```python
params = {'batch_size':[10,20,30],
          'nb_epoch':[20,50],
          'optimizer':['adam','rmsprop']
         }
```

```python
import sklearn.model_selection as skmod
```

```python
gs=skmod.GridSearchCV(mod2,params,scoring='accuracy',cv=10)
```

```python
gs = gs.fit(X,Y)
```
